```python
# Using horizontal bar chart to visualize the count and percentage of the job link
urls = df['job_apply_link']

# Extract the site names from the URLs
site_names1 = [urlparse(url).netloc for url in urls]

# Convert the list to a pandas Series and get the unique counts
unique_site_counts = pd.Series(site_names1).value_counts()

# Sort the unique counts in descending order
unique_site_counts_sorted = unique_site_counts.sort_values(ascending=False)

# Calculate the percentage of each site's contribution to the total
percentages = (unique_site_counts_sorted / unique_site_counts_sorted.sum()) * 100

# Plot horizontal bar chart with annotations for percentages
plt.figure(figsize=(10, 6))
ax = unique_site_counts_sorted.plot(kind='barh', color='cyan')
plt.xlabel('Count')
plt.ylabel('Site Name')
plt.title('Unique Counts of Each Site Name')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest count at the top

# Annotate each bar with its percentage contribution
for i, v in enumerate(unique_site_counts_sorted):
    ax.text(v + 0.5, i, f'{v} ({percentages.iloc[i]:.1f}%)', va='center')

plt.xlim(right=max(unique_site_counts_sorted) + 10)  # Set xlim with some padding

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

## PLOTLY

```python
import plotly.graph_objs as go

# Drop rows with missing values in 'job_title' and 'job_city' columns
df.dropna(subset=['job_title', 'job_city'], inplace=True)

# Count the occurrences of each job title in each city
job_demand = df.groupby(['job_title', 'job_city']).size().reset_index(name='count')

# Sort the dataframe by count in descending order to focus on the most in-demand titles
job_demand.sort_values(by='count', ascending=False, inplace=True)

# Select the top N job titles for visualization (e.g., top 10)
top_n_titles = job_demand['job_title'].unique()[:15]

# Filter the dataframe to include only the top job titles
job_demand_top = job_demand[job_demand['job_title'].isin(top_n_titles)]

# Pivot the dataframe to prepare for plotting
pivot_job_demand = job_demand_top.pivot(index='job_title', columns='job_city', values='count')

# Define data traces for each city
data = []
for city in pivot_job_demand.columns:
    trace = go.Bar(
        x=pivot_job_demand.index,
        y=pivot_job_demand[city],
        name=city,
        hovertemplate='%{y} job postings in %{x} in ' + city
    )
    data.append(trace)

# Define layout with increased width and height
layout = go.Layout(
    title='Job Demand by Title and Location',
    xaxis=dict(title='Job Title', tickangle=-45),
    yaxis=dict(title='Number of Job Postings'),
    barmode='stack',
    width=1000,  # Adjust width as needed
    height=1000   # Adjust height as needed
)

# Create figure
fig = go.Figure(data=data, layout=layout)

# Show figure
fig.show()
```
